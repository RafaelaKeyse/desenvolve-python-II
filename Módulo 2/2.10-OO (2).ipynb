* A classe **não possui construtor customizado**. A alteração de seu atributo se dará a partir da propriedade a seguir.
* Crie a `property` `data_hora` para manipular o atributo `_data_hora`.
    * O getter da propriedade deve retornar a data como uma string formatada (`%d/%m/%Y, %H:%M`). Use o atributo `FORMAT`. Consulte o [funcionamento do método `strftime`](https://www.programiz.com/python-programming/datetime/strftime).
    * O setter da propriedade deve receber uma string de data formatada (`%d/%m/%Y, %H:%M`) e implementar um bloco try-except que tenta converter a string em `datetime` e lança um `ValueError` caso a entrada seja inválida. Use o atributo `FORMAT`. Consulte o [funcionamento do método `strptime`](https://www.digitalocean.com/community/tutorials/python-string-to-datetime-strptime).
* Crie o método de instância `isPassado(self)` que avalia se a `_data_hora` é menor que `datetime.now()` (a data e hora atual) e retorna `True` em caso positivo, e `False` caso contrário.
* Crie o método de instância `somaDias(self, num_dias)` que recebe um inteiro `num_dias`, soma esse valor ao atributo interno `_data_hora` e retorna a string formatada do resultado da soma (código dado a seguir).

Teste a classe `DataHora` com o seguinte código (altere o que for necessário):
```python
# instanciando o objeto
dh = DataHora()

# definindo a data_hora através da propriedade
dh.data_hora = '05/02/2024, 12:30'

## editando a data_hora através da função somaDias
dh.data_hora = dh.somaDias(30)

## imprimindo a data_hora editada e se é passado
print(dh.data_hora, dh.isPassado())
```

````python
#### Classe DataHora
from datetime import datetime, timedelta

class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'  # atributo de classe

    def __init__(self):
        self._data_hora = None  # atributo privado

    @property
    def data_hora(self):
        if self._data_hora:
            return self._data_hora.strftime(self.FORMAT)
        return None

    @data_hora.setter
    def data_hora(self, valor_str):
        try:
            self._data_hora = datetime.strptime(valor_str, self.FORMAT)
        except ValueError:
            raise ValueError(f"Data inválida. Use o formato {self.FORMAT}")

    def isPassado(self):
        if self._data_hora is None:
            return False
        return self._data_hora < datetime.now()

    def somaDias(self, num_dias):
        if self._data_hora is None:
            raise ValueError("Data não definida.")
        data_hora_somada = self._data_hora + timedelta(days=num_dias)
        self._data_hora = data_hora_somada
        return self._data_hora.strftime(self.FORMAT)

# Teste da classe DataHora
dh = DataHora()
dh.data_hora = '05/02/2024, 12:30'
print("Original:", dh.data_hora, dh.isPassado())  # imprime data original e se é passado

dh.data_hora = dh.somaDias(30)  # soma 30 dias e atualiza a data
print("Após somaDias:", dh.data_hora, dh.isPassado())  # imprime nova data e se é passado
````

````python
#### Classe EventoUnico
from datetime import datetime

# Supondo que EventoABC seja uma classe abstrata com os atributos _titulo e _descricao
from abc import ABC, abstractmethod

class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def isConcluido(self):
        pass

    @abstractmethod
    def __str__(self):
        pass

# Classe auxiliar DataHora
class DataHora:
    def __init__(self, data_hora_str):
        self._data_hora = datetime.strptime(data_hora_str, '%d/%m/%Y, %H:%M')

    @property
    def data_hora(self):
        return self._data_hora

    @data_hora.setter
    def data_hora(self, nova_data_str):
        self._data_hora = datetime.strptime(nova_data_str, '%d/%m/%Y, %H:%M')

    def isPassado(self):
        return datetime.now() > self._data_hora

    def __str__(self):
        return self._data_hora.strftime('%d/%m/%Y, %H:%M')

# Classe EventoUnico
class EventoUnico(EventoABC):
    def __init__(self, titulo, descricao, data_hora_str):
        super().__init__(titulo, descricao)
        self._data_hora = DataHora(data_hora_str)

    def isConcluido(self):
        return self._data_hora.isPassado()

    def __str__(self):
        return f"Evento: {self._titulo}, Data: {self._data_hora}, Descrição: {self._descricao}, Concluido: {self.isConcluido()}"

    def editar_data_hora(self, nova_data_str):
        self._data_hora.data_hora = nova_data_str

# Testando a classe
evento = EventoUnico('Reunião', 'Sala 302, prédio da esquina', '05/10/2023, 16:30')
print(evento)

evento.editar_data_hora('05/10/2024, 16:30')
print(evento)
````

Crie a classe `EventoRecorrente`:
* A classe deve herdar de `EventoABC`.
* Possui como atributo próprio uma lista privada de objetos `DataHora` (como você deve nomear o atributo?).
* Seu construtor recebe os atributos da superclasse, além dos atributos `data_hora_inicial` (string formatada), `data_hora_final` (string formatada) e `intervalo_repeticao` (int), sendo o intervalo dado em dias. Preencha a coleção `DataHora` de acordo com o intervalo de repetição fornecido. Dica: crie o objeto `DataHora` inicial e use sua função interna `somaDias` para criar iterativamente as novas instâncias do intervalo até chegar em `DataHora` final. 
*  Implementa os métodos abstratos da superclasse:
    * Método `isConcluido(indice)` que que invoca o método `isPassado()` do elemento `indice` da coleção de objetos `DataHora` e retorna seu resultado. 
    * Método `__str__` que imprime (em um laço) **todos as ocorrências `i` do evento** na forma `"Evento: _titulo, Data: data_hora[i], Descrição: _descricao, Concluido: isConcluido(i)"`. 
* Crie o método `editar_data_hora` que recebe `data_hora_antiga` e `data_hora_nova` e altera o elemento da coleção de objetos `DataHora` que corresponde a `data_hora_antiga` fornecida.    


Teste a classe `EventoRecorrente` com o seguinte código:
```python
# criar evento

    'Reunião', 'Sala 302, prédio da esquina', 
    '05/01/2024, 16:30', '05/01/2025, 16:30', 30)

print(eventos)

eventos.editar_data_hora('05/12/2024, 16:30', '05/12/2024, 11:30')
```

````python
from abc import ABC, abstractmethod
from datetime import datetime, timedelta

# Classe abstrata EventoABC
class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def isConcluido(self, indice=None):
        pass

    @abstractmethod
    def __str__(self):
        pass

# Classe DataHora com suporte à somaDias
class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'

    def __init__(self, data_hora_str):
        self.data_hora = data_hora_str  # usa o setter

    @property
    def data_hora(self):
        return self._data_hora

    @data_hora.setter
    def data_hora(self, nova_data_str):
        try:
            self._data_hora = datetime.strptime(nova_data_str, self.FORMAT)
        except ValueError:
            raise ValueError(f"Data '{nova_data_str}' não está no formato {self.FORMAT}")

    def isPassado(self):
        return datetime.now() > self._data_hora

    def somaDias(self, num_dias):
        data_somada = self._data_hora + timedelta(days=num_dias)
        return data_somada.strftime(self.FORMAT)

    def __str__(self):
        return self._data_hora.strftime(self.FORMAT)

# Classe EventoRecorrente
class EventoRecorrente(EventoABC):
    def __init__(self, titulo, descricao, data_hora_inicial_str, data_hora_final_str, intervalo_repeticao_dias):
        super().__init__(titulo, descricao)
        self._datas = []  # lista de objetos DataHora

        data_atual = DataHora(data_hora_inicial_str)
        data_final = datetime.strptime(data_hora_final_str, DataHora.FORMAT)

        while data_atual.data_hora <= data_final:
            self._datas.append(data_atual)
            nova_data_str = data_atual.somaDias(intervalo_repeticao_dias)
            data_atual = DataHora(nova_data_str)

    def isConcluido(self, indice):
        if indice < 0 or indice >= len(self._datas):
            raise IndexError("Índice fora do intervalo dos eventos.")
        return self._datas[indice].isPassado()

    def __str__(self):
        resultado = ""
        for i, data_obj in enumerate(self._datas):
            resultado += (f"Evento: {self._titulo}, Data: {data_obj}, "
                          f"Descrição: {self._descricao}, Concluido: {self.isConcluido(i)}\n")
        return resultado.strip()

    def editar_data_hora(self, data_hora_antiga_str, data_hora_nova_str):
        # Certifique-se que a string nova tem o espaço depois da vírgula
        if ', ' not in data_hora_nova_str:
            data_hora_nova_str = data_hora_nova_str.replace(',', ', ')

        data_antiga_obj = datetime.strptime(data_hora_antiga_str, DataHora.FORMAT)
        for i, data_obj in enumerate(self._datas):
            if data_obj.data_hora == data_antiga_obj:
                self._datas[i].data_hora = data_hora_nova_str
                return
        raise ValueError(f"Data {data_hora_antiga_str} não encontrada entre os eventos.")

# --- Teste da classe ---

eventos = EventoRecorrente(
    'Reunião', 'Sala 302, prédio da esquina',
    '05/01/2024, 16:30', '05/01/2025, 16:30', 30
)

print("Eventos gerados:")
print(eventos)

print("\nDatas disponíveis para edição:")
for data_obj in eventos._datas:
    print(data_obj)

print("\nEditando evento na data: 31/10/2024, 16:30...")
# Atenção: tem que passar exatamente a data como string da lista, incluindo o espaço após a vírgula
eventos.editar_data_hora('31/10/2024, 16:30', '31/10/2024, 11:30')

print("\nEventos após edição:")
print(eventos)
````

Por fim, vamos só ver o polimorfismo em ação. Crie e preencha uma lista de eventos, sendo alguns do tipo `EventoUnico` e outros do tipo `EventoRecorrente`. Sobre essa lista, execute o laço de impressão a seguir:
```python
for evento in lista_eventos: print(evento)
```
A função `print` irá invocar o método especial `__str__` das classes correspondentes dependendo do tipo do objeto recebido. Aí está o polimorfismo :)

````python
from abc import ABC, abstractmethod
from datetime import datetime, timedelta

# Classe abstrata
class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def isConcluido(self, indice=None):
        pass

    @abstractmethod
    def __str__(self):
        pass

# Classe auxiliar
class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'

    def __init__(self, data_hora_str):
        self.data_hora = data_hora_str  # usa o setter

    @property
    def data_hora(self):
        return self._data_hora

    @data_hora.setter
    def data_hora(self, nova_data_str):
        try:
            self._data_hora = datetime.strptime(nova_data_str, self.FORMAT)
        except ValueError:
            raise ValueError(f"Data '{nova_data_str}' não está no formato {self.FORMAT}")

    def isPassado(self):
        return datetime.now() > self._data_hora

    def somaDias(self, num_dias):
        data_somada = self._data_hora + timedelta(days=num_dias)
        return data_somada.strftime(self.FORMAT)

    def __str__(self):
        return self._data_hora.strftime(self.FORMAT)

# Evento único
class EventoUnico(EventoABC):
    def __init__(self, titulo, descricao, data_hora_str):
        super().__init__(titulo, descricao)
        self._data_hora = DataHora(data_hora_str)

    def isConcluido(self, indice=None):
        return self._data_hora.isPassado()

    def editar_data_hora(self, nova_data_str):
        self._data_hora.data_hora = nova_data_str

    def __str__(self):
        return (f"Evento: {self._titulo}, Data: {self._data_hora}, "
                f"Descrição: {self._descricao}, Concluido: {self.isConcluido()}")

# Evento recorrente
class EventoRecorrente(EventoABC):
    def __init__(self, titulo, descricao, data_hora_inicial_str, data_hora_final_str, intervalo_repeticao_dias):
        super().__init__(titulo, descricao)
        self._datas = []

        data_atual = DataHora(data_hora_inicial_str)
        data_final = datetime.strptime(data_hora_final_str, DataHora.FORMAT)

        while data_atual.data_hora <= data_final:
            self._datas.append(data_atual)
            nova_data_str = data_atual.somaDias(intervalo_repeticao_dias)
            data_atual = DataHora(nova_data_str)

    def isConcluido(self, indice):
        if indice < 0 or indice >= len(self._datas):
            raise IndexError("Índice fora do intervalo dos eventos.")
        return self._datas[indice].isPassado()

    def editar_data_hora(self, data_hora_antiga_str, data_hora_nova_str):
        if ', ' not in data_hora_nova_str:
            data_hora_nova_str = data_hora_nova_str.replace(',', ', ')
        data_antiga_obj = datetime.strptime(data_hora_antiga_str, DataHora.FORMAT)
        for i, data_obj in enumerate(self._datas):
            if data_obj.data_hora == data_antiga_obj:
                self._datas[i].data_hora = data_hora_nova_str
                return
        raise ValueError(f"Data {data_hora_antiga_str} não encontrada entre os eventos.")

    def __str__(self):
        resultado = ""
        for i, data_obj in enumerate(self._datas):
            resultado += (f"Evento: {self._titulo}, Data: {data_obj}, "
                          f"Descrição: {self._descricao}, Concluido: {self.isConcluido(i)}\n")
        return resultado.strip()

# Demonstração do polimorfismo:
evento1 = EventoUnico('Consulta Médica', 'Clínica Vida', '10/07/2024, 09:00')
evento2 = EventoUnico('Palestra', 'Auditório Central', '20/07/2024, 14:00')
evento3 = EventoRecorrente('Reunião Mensal', 'Sala 302', '01/01/2024, 10:00', '01/04/2024, 10:00', 30)
evento4 = EventoRecorrente('Aula de Python', 'Lab 5', '05/01/2024, 08:00', '05/03/2024, 08:00', 15)

lista_eventos = [evento1, evento2, evento3, evento4]

print("\n📌 Lista de Eventos (polimorfismo em ação):\n")
for evento in lista_eventos:
    print(evento)
    print('-' * 80)
````
