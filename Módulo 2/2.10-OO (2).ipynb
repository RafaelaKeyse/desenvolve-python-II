Teste a classe `DataHora` com o seguinte c√≥digo (altere o que for necess√°rio):
```python
# instanciando o objeto

# definindo a data_hora atrav√©s da propriedade
dh.data_hora = '05/02/2024, 12:30'

## editando a data_hora atrav√©s da fun√ß√£o somaDias
```

````python
#### Classe DataHora
from datetime import datetime, timedelta

class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'  # atributo de classe

    def __init__(self):
        self._data_hora = None  # atributo privado

    @property
    def data_hora(self):
        if self._data_hora:
            return self._data_hora.strftime(self.FORMAT)
        return None

    @data_hora.setter
    def data_hora(self, valor_str):
        try:
            self._data_hora = datetime.strptime(valor_str, self.FORMAT)
        except ValueError:
            raise ValueError(f"Data inv√°lida. Use o formato {self.FORMAT}")

    def isPassado(self):
        if self._data_hora is None:
            return False
        return self._data_hora < datetime.now()

    def somaDias(self, num_dias):
        if self._data_hora is None:
            raise ValueError("Data n√£o definida.")
        data_hora_somada = self._data_hora + timedelta(days=num_dias)
        self._data_hora = data_hora_somada
        return self._data_hora.strftime(self.FORMAT)

# Teste da classe DataHora
dh = DataHora()
dh.data_hora = '05/02/2024, 12:30'
print("Original:", dh.data_hora, dh.isPassado())  # imprime data original e se √© passado

dh.data_hora = dh.somaDias(30)  # soma 30 dias e atualiza a data
print("Ap√≥s somaDias:", dh.data_hora, dh.isPassado())  # imprime nova data e se √© passado
````

````python
#### Classe EventoUnico
from datetime import datetime

# Supondo que EventoABC seja uma classe abstrata com os atributos _titulo e _descricao
from abc import ABC, abstractmethod

class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def isConcluido(self):
        pass

    @abstractmethod
    def __str__(self):
        pass

# Classe auxiliar DataHora
class DataHora:
    def __init__(self, data_hora_str):
        self._data_hora = datetime.strptime(data_hora_str, '%d/%m/%Y, %H:%M')

    @property
    def data_hora(self):
        return self._data_hora

    @data_hora.setter
    def data_hora(self, nova_data_str):
        self._data_hora = datetime.strptime(nova_data_str, '%d/%m/%Y, %H:%M')

    def isPassado(self):
        return datetime.now() > self._data_hora

    def __str__(self):
        return self._data_hora.strftime('%d/%m/%Y, %H:%M')

# Classe EventoUnico
class EventoUnico(EventoABC):
    def __init__(self, titulo, descricao, data_hora_str):
        super().__init__(titulo, descricao)
        self._data_hora = DataHora(data_hora_str)

    def isConcluido(self):
        return self._data_hora.isPassado()

    def __str__(self):
        return f"Evento: {self._titulo}, Data: {self._data_hora}, Descri√ß√£o: {self._descricao}, Concluido: {self.isConcluido()}"

    def editar_data_hora(self, nova_data_str):
        self._data_hora.data_hora = nova_data_str

# Testando a classe
evento = EventoUnico('Reuni√£o', 'Sala 302, pr√©dio da esquina', '05/10/2023, 16:30')
print(evento)

evento.editar_data_hora('05/10/2024, 16:30')
print(evento)
````

* Seu construtor recebe os atributos da superclasse, al√©m dos atributos `data_hora_inicial` (string formatada), `data_hora_final` (string formatada) e `intervalo_repeticao` (int), sendo o intervalo dado em dias.     * M√©todo `__str__` que imprime (em um la√ßo) **todos as ocorr√™ncias `i` do evento** na forma `"Evento: _titulo, Data: data_hora[i], Descri√ß√£o: _descricao, Concluido: isConcluido(i)"`.

Teste a classe `EventoRecorrente` com o seguinte c√≥digo:

````python
from abc import ABC, abstractmethod
from datetime import datetime, timedelta

# Classe abstrata EventoABC
class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def isConcluido(self, indice=None):
        pass

    @abstractmethod
    def __str__(self):
        pass

# Classe DataHora com suporte √† somaDias
class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'

    def __init__(self, data_hora_str):
        self.data_hora = data_hora_str  # usa o setter

    @property
    def data_hora(self):
        return self._data_hora

    @data_hora.setter
    def data_hora(self, nova_data_str):
        try:
            self._data_hora = datetime.strptime(nova_data_str, self.FORMAT)
        except ValueError:
            raise ValueError(f"Data '{nova_data_str}' n√£o est√° no formato {self.FORMAT}")

    def isPassado(self):
        return datetime.now() > self._data_hora

    def somaDias(self, num_dias):
        data_somada = self._data_hora + timedelta(days=num_dias)
        return data_somada.strftime(self.FORMAT)

    def __str__(self):
        return self._data_hora.strftime(self.FORMAT)

# Classe EventoRecorrente
class EventoRecorrente(EventoABC):
    def __init__(self, titulo, descricao, data_hora_inicial_str, data_hora_final_str, intervalo_repeticao_dias):
        super().__init__(titulo, descricao)
        self._datas = []  # lista de objetos DataHora

        data_atual = DataHora(data_hora_inicial_str)
        data_final = datetime.strptime(data_hora_final_str, DataHora.FORMAT)

        while data_atual.data_hora <= data_final:
            self._datas.append(data_atual)
            nova_data_str = data_atual.somaDias(intervalo_repeticao_dias)
            data_atual = DataHora(nova_data_str)

    def isConcluido(self, indice):
        if indice < 0 or indice >= len(self._datas):
            raise IndexError("√çndice fora do intervalo dos eventos.")
        return self._datas[indice].isPassado()

    def __str__(self):
        resultado = ""
        for i, data_obj in enumerate(self._datas):
            resultado += (f"Evento: {self._titulo}, Data: {data_obj}, "
                          f"Descri√ß√£o: {self._descricao}, Concluido: {self.isConcluido(i)}\n")
        return resultado.strip()

    def editar_data_hora(self, data_hora_antiga_str, data_hora_nova_str):
        # Certifique-se que a string nova tem o espa√ßo depois da v√≠rgula
        if ', ' not in data_hora_nova_str:
            data_hora_nova_str = data_hora_nova_str.replace(',', ', ')

        data_antiga_obj = datetime.strptime(data_hora_antiga_str, DataHora.FORMAT)
        for i, data_obj in enumerate(self._datas):
            if data_obj.data_hora == data_antiga_obj:
                self._datas[i].data_hora = data_hora_nova_str
                return
        raise ValueError(f"Data {data_hora_antiga_str} n√£o encontrada entre os eventos.")

# --- Teste da classe ---

eventos = EventoRecorrente(
    'Reuni√£o', 'Sala 302, pr√©dio da esquina',
    '05/01/2024, 16:30', '05/01/2025, 16:30', 30
)

print("Eventos gerados:")
print(eventos)

print("\nDatas dispon√≠veis para edi√ß√£o:")
for data_obj in eventos._datas:
    print(data_obj)

print("\nEditando evento na data: 31/10/2024, 16:30...")
# Aten√ß√£o: tem que passar exatamente a data como string da lista, incluindo o espa√ßo ap√≥s a v√≠rgula
eventos.editar_data_hora('31/10/2024, 16:30', '31/10/2024, 11:30')

print("\nEventos ap√≥s edi√ß√£o:")
print(eventos)
````

Por fim, vamos s√≥ ver o polimorfismo em a√ß√£o. Crie e preencha uma lista de eventos, sendo alguns do tipo `EventoUnico` e outros do tipo `EventoRecorrente`. Sobre essa lista, execute o la√ßo de impress√£o a seguir:

A fun√ß√£o `print` ir√° invocar o m√©todo especial `__str__` das classes correspondentes dependendo do tipo do objeto recebido. A√≠ est√° o polimorfismo :)

````python
from abc import ABC, abstractmethod
from datetime import datetime, timedelta

# Classe abstrata
class EventoABC(ABC):
    def __init__(self, titulo, descricao):
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def isConcluido(self, indice=None):
        pass

    @abstractmethod
    def __str__(self):
        pass

# Classe auxiliar
class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'

    def __init__(self, data_hora_str):
        self.data_hora = data_hora_str  # usa o setter

    @property
    def data_hora(self):
        return self._data_hora

    @data_hora.setter
    def data_hora(self, nova_data_str):
        try:
            self._data_hora = datetime.strptime(nova_data_str, self.FORMAT)
        except ValueError:
            raise ValueError(f"Data '{nova_data_str}' n√£o est√° no formato {self.FORMAT}")

    def isPassado(self):
        return datetime.now() > self._data_hora

    def somaDias(self, num_dias):
        data_somada = self._data_hora + timedelta(days=num_dias)
        return data_somada.strftime(self.FORMAT)

    def __str__(self):
        return self._data_hora.strftime(self.FORMAT)

# Evento √∫nico
class EventoUnico(EventoABC):
    def __init__(self, titulo, descricao, data_hora_str):
        super().__init__(titulo, descricao)
        self._data_hora = DataHora(data_hora_str)

    def isConcluido(self, indice=None):
        return self._data_hora.isPassado()

    def editar_data_hora(self, nova_data_str):
        self._data_hora.data_hora = nova_data_str

    def __str__(self):
        return (f"Evento: {self._titulo}, Data: {self._data_hora}, "
                f"Descri√ß√£o: {self._descricao}, Concluido: {self.isConcluido()}")

# Evento recorrente
class EventoRecorrente(EventoABC):
    def __init__(self, titulo, descricao, data_hora_inicial_str, data_hora_final_str, intervalo_repeticao_dias):
        super().__init__(titulo, descricao)
        self._datas = []

        data_atual = DataHora(data_hora_inicial_str)
        data_final = datetime.strptime(data_hora_final_str, DataHora.FORMAT)

        while data_atual.data_hora <= data_final:
            self._datas.append(data_atual)
            nova_data_str = data_atual.somaDias(intervalo_repeticao_dias)
            data_atual = DataHora(nova_data_str)

    def isConcluido(self, indice):
        if indice < 0 or indice >= len(self._datas):
            raise IndexError("√çndice fora do intervalo dos eventos.")
        return self._datas[indice].isPassado()

    def editar_data_hora(self, data_hora_antiga_str, data_hora_nova_str):
        if ', ' not in data_hora_nova_str:
            data_hora_nova_str = data_hora_nova_str.replace(',', ', ')
        data_antiga_obj = datetime.strptime(data_hora_antiga_str, DataHora.FORMAT)
        for i, data_obj in enumerate(self._datas):
            if data_obj.data_hora == data_antiga_obj:
                self._datas[i].data_hora = data_hora_nova_str
                return
        raise ValueError(f"Data {data_hora_antiga_str} n√£o encontrada entre os eventos.")

    def __str__(self):
        resultado = ""
        for i, data_obj in enumerate(self._datas):
            resultado += (f"Evento: {self._titulo}, Data: {data_obj}, "
                          f"Descri√ß√£o: {self._descricao}, Concluido: {self.isConcluido(i)}\n")
        return resultado.strip()

# Demonstra√ß√£o do polimorfismo:
evento1 = EventoUnico('Consulta M√©dica', 'Cl√≠nica Vida', '10/07/2024, 09:00')
evento2 = EventoUnico('Palestra', 'Audit√≥rio Central', '20/07/2024, 14:00')
evento3 = EventoRecorrente('Reuni√£o Mensal', 'Sala 302', '01/01/2024, 10:00', '01/04/2024, 10:00', 30)
evento4 = EventoRecorrente('Aula de Python', 'Lab 5', '05/01/2024, 08:00', '05/03/2024, 08:00', 15)

lista_eventos = [evento1, evento2, evento3, evento4]

print("\nüìå Lista de Eventos (polimorfismo em a√ß√£o):\n")
for evento in lista_eventos:
    print(evento)
    print('-' * 80)
````
